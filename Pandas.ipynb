```python
nComps = energyDF["Company Name"].nunique()
```

```python
print (nComps)
```

```python
queensJobs = energyDF[energyDF["City"] == "Queens"]["Job created"].sum()
print (queensJobs)
```

```python
emailsDF = energyDF.copy().dropna(subset = ["company email"])
emailsDF.info()
```

```python
def getDomain(email):
    emailList = email.split("@")
    if len(emailList) != 2:
        # Filter out incorrectly input emails
        # print (email)
        return pd.NaT
    else:
        return emailList[1]
```

```python
emailsDF["domain"] = emailsDF["company email"].apply(lambda x : getDomain(x))
uniqueDomains = emailsDF["domain"].nunique()
print (uniqueDomains)
```

```python
NTAs = energyDF.groupby("Neighborhood Tabulation Area (NTA) (2020)").filter(lambda x : len(x) >= 5)
```

```python
avgJobsCreated = NTAs[NTAs["Job created"].notnull()]["Job created"].mean()
print (avgJobsCreated)
```

```python
NTAs["Total Savings"]
```

```python
avgTotalSavings = NTAs["Total Savings"].mean()
print (avgTotalSavings)
```

```python
import csv
with open("./energyResults.csv", 'w') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(["UniqueCompanies","total jobs created in Queens","unique email domains","avg total savings","avg total jobs created"])
    csvwriter.writerow([nComps,queensJobs, uniqueDomains, avgJobsCreated, avgTotalSavings])
```
